```python
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
import re
import pandas as pd
```

## Web spider

```python
def instructor_links(driver,url): ### FOR cpe.kmutt.ac.th instructor -th/-en website
    FROM = driver.get(url)
    postid = driver.find_elements_by_css_selector("a[postid]")
    for link in range(len(postid)):
        instructor = driver.find_elements_by_css_selector("a[postid]")
        urls_from.append(str(driver.current_url))
        instructor[link].click()
        urls_to.append(str(driver.current_url))
        sleep(0.05)
        driver.back()
```

```python
def blog_links(driver,url,post_url):
    blog = driver.get(url)
    page = driver.find_elements_by_css_selector("a[id]")
    for number in range(1,len(page)):
        page_no = driver.find_element_by_id(str(number))
        page_no.click()
        soup = BeautifulSoup(driver.page_source,'html.parser')
        LINKS = soup.select('a[postid]')
        for link in LINKS:
            urls_from.append(str(driver.current_url))
            postid = link.get("postid")
            urls_to.append(post_url+postid)
        sleep(0.5)
    driver.close()
```

```python
######## SETUP DRIVER
driver = webdriver.Edge(executable_path='./msedgedriver')
driver.implicitly_wait(30)


# List urls
urls_from = []
urls_to = []  

# Scraping
CPE_site = "https://www.cpe.kmutt.ac.th"
driver.get(CPE_site)
soup = BeautifulSoup(driver.page_source,'html.parser')
href = soup.select("a[href^='/']")

#Main website TH Language
for link in href:
    tmp = link.get('href')
    url_from = driver.current_url
    urls_from.append(url_from)
    urls_to.append(CPE_site+tmp)
    
    ### travasal
    driver.get(CPE_site+tmp)
    sleep(0.5)
    
    if str(driver.current_url) == "https://www.cpe.kmutt.ac.th/th/staff": ### Staff collecting url th
        sleep(0.05)
        instructor_links(driver,str(driver.current_url))
                
### ENGLISH LANGUAGE Main website scraping
driver.get(CPE_site)
eng_ver =driver.find_element_by_id("thaiWeb")
eng_ver.click()
soup_en = BeautifulSoup(driver.page_source, 'html.parser')
href_en = soup_en.select("a[href^='/']")

for link in href_en:
    tmp = link.get('href')
    url_from = driver.current_url
    urls_from.append(url_from)
    urls_to.append(CPE_site+tmp)
    
    ### travasal
    driver.get(CPE_site+tmp)
    sleep(0.5)
    
    if str(driver.current_url) == "https://www.cpe.kmutt.ac.th/en/staff": ### Staff collecting url en
        sleep(0.05)
        instructor_links(driver,str(driver.current_url))
        
driver.close()

```

```python
# Blog cpe website travasal links TH
driver = webdriver.Edge(executable_path='./msedgedriver')
driver.implicitly_wait(30)

blog_from_TH = 'https://www.cpe.kmutt.ac.th/th/blog'
blog_post_TH = 'https://www.cpe.kmutt.ac.th/th/post/'

blog_links(driver,blog_from_TH,blog_post_TH)
```

```python
# Blog cpe website travasal links EN
driver = webdriver.Edge(executable_path='./msedgedriver')
driver.implicitly_wait(30)

blog_from_EN = 'https://www.cpe.kmutt.ac.th/en/blog'
blog_post_EN = 'https://www.cpe.kmutt.ac.th/en/post/'

blog_links(driver,blog_from_EN,blog_post_EN)
```

```python
title_main = []
title_list = []
content_list = []

driver = webdriver.Edge(executable_path='./msedgedriver')
driver.implicitly_wait(30)

for link in cpe_urls:
    driver.get(link)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    
    if str(links) == 'https://www.cpe.kmutt.ac.th/th/index': ### main website TH
        sub_titlelist = []
        title = soup.select("h3")
        title_main.apppend(title[0].get_text())
        for subtitle in range(1, len(title)): # index 0 for main
            sub_titlelist.append(title[subtitle].get_text())
        title_list.append(sub_titlelist)
        continue
        
    elif str(links) == 'https://www.cpe.kmutt.ac.th/en/index': ### main website EN
        sub_titlelist = []
        title = soup.select("h3")
        title_main.apppend(title[0].get_text())
        for subtitle in range(1, len(title)): # index 0 for main
            sub_titlelist.append(title[subtitle].get_text())
        title_list.append(sub_titlelist)
        continue

##### check and load MAIN TITLE

    main_title = soup.select("h1")
    if len(main_title) != 0:
        title_main.append(main_title[0].get_text())
    else:
        title_main.append("None")
    
    sub_title = soup.select("h3.box-title")
    list_unit = []
    text_content = []
    text_bullet = []

#### check and load SUBTITLE

    if len(sub_title) != 0 :
        for unit in range(len(sub_title)):
            list_unit.append(sub_title[unit].get_text())
        title_list.append(list_unit)
    else:
        title_list.append("None")
        
#### check and load CONTENT


    staff_info = soup.select("p.staff-list-detail")
    content = soup.select("div > p")
    content_bullet = soup.select("div.col-xs-12 > ul")
    
    if len(content) != 0:
        for text in range(len(content)):
            text_content.append(content[text].get_text())
            
    if len(staff_info) != 0:
        for info in range(len(staff_info)):
            text_content.append(staff_info[info].get_text())
        
#### check and load BULLET CONTENT         
        if len(content_bullet) != 0:
            for text in range(len(content_bullet)):
                text_bullet.append(content_bullet[text].get_text())
            text_content.append(text_bullet)
            
    content_list.append(text_content)  
    

        
    sleep(0.05)

    
        
        
```
